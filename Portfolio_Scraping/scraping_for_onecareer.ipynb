```python
import time
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import csv
import chromedriver_binary
```

```python
def get_driver():
    options = Options()
    options.add_argument('--headless')
    options.add_argument("--no-sandbox")
    return webdriver.Chrome( options = options )

def get_page_bs( driver, url ):
    driver.get( url )
    time.sleep( 10 ) # 10秒間待機
    return BeautifulSoup( driver.page_source, features='lxml' )

def write_csv( file_path, data ):
    with open( file_path, 'a' ) as f:
        print( data, file = f )
        
def write_csv_list( file_path, data ):
    with open( file_path, 'a' ) as f:
        print( ','.join( data ), file=f )

def get_company_detail( driver, url, output_file ):
    bs = get_page_bs( driver, url )
    div_tags = bs.find_all( 'div', class_ = 'v2-companies__top-middle' )
    print( "＊＊＊ " + url + " のデータを取得中＊＊＊")

    for item in div_tags:
        name = item.find( 'a', class_ = 'v2-companies__title' )
        field = item.find( 'div', class_ = 'v2-companies__business-field' )
        print( "データ取得中...:", name.text.replace( '\n', '' ), field.find_all( 'span' )[1].text.replace( '\n', '' ) )
        data = ( name.text.replace( '\n', '' ), field.find_all( 'span' )[1].text.replace( '\n', '' ) )
        write_csv_list( output_file, data )

## データ取得をする ページURL / データを保存するファイルパス は適宜変更する ###
target_url = "https://www.onecareer.jp/companies/business_categories/3?page=1"
output_file_path = 'data/onecareer_company_list_1to40.csv'

d = get_driver()
write_csv( output_file_path, '企業名,分野' )
for i in range(41,367):
    get_company_detail( d, target_url[:len(target_url)-1] + str(i), output_file_path )
```
