```python
import time
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import csv
import chromedriver_binary
import urllib
import datetime

## データ取得をする ページURL / データを保存するファイルパス は適宜変更する ##
output = "prtimes_data.csv"
```

```python
def get_driver():
    options = Options()
    options.add_argument('--headless')
    options.add_argument("--no-sandbox")
    return webdriver.Chrome( options = options )

def get_page_bs( driver, url ):
    driver.get( url )
    # time.sleep( 5 ) # 10秒間待機
    return BeautifulSoup( driver.page_source, features='lxml' )
        
def write_csv( file_path, data ):
    with open( file_path, 'a' ) as f:
        print( ','.join( data ), file=f )
    
d = get_driver()
bs = get_page_bs( d, "file:///Users/kt/Downloads/prtimes8000.html" )
articles = bs.find_all( 'article', class_ = 'list-article' )
print( len( articles ) )
write_csv( output, ("企業名","記事名","記事URL","PRTIMES企業紹介URL","公開日","公開時間") )

for i in range(0,9400):
    article_url = articles[ i ].find( 'a', class_ = 'list-article__link' ).get('href')
    title = articles[ i ].find( 'h3', class_= 'list-article__title' ).text.replace(" ","").replace("\n","")
    date = articles[ i ].find( 'time' ).get('datetime').replace(" ","").replace("\n","")[:10]
    time = articles[ i ].find( 'time' ).get('datetime').replace(" ","").replace("\n","")[11:19]
    d_s = date.split('-')
    t_s = time.split(':')
    date_time = datetime.datetime( int(d_s[0]), int(d_s[1]), int(d_s[2]), int(t_s[0]), int(t_s[1]), int(t_s[2]) ) + datetime.timedelta(hours=9)
    date_f = "{0:%Y/%m/%d}".format(date_time)
    time_f = "{0:%H:%M:%S}".format(date_time)
    company_name = articles[ i ].find( 'a', class_ = 'list-article__company-name-link' ).text.replace(" ","").replace("\n","")
    company_url = articles[ i ].find( 'a', class_ = 'list-article__company-name-link' ).get('href')
    print( article_url, title, date_f, time_f,company_name,company_url )
    write_csv( output, (company_name, title, article_url, company_url, date_f, time_f) )
```
