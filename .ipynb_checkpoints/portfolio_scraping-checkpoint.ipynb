```python
from time import sleep
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import textwrap
import csv
import chromedriver_binary
import time
```

```python
# URL を与えると， その Webページの HTML を返す
def get_page( url ):
    options = Options()
    options.add_argument('--headless')
    options.add_argument("--no-sandbox")
    driver = webdriver.Chrome(options=options)
    #driver = webdriver.Chrome()
    driver.get( url )
    time.sleep(5)
    res = driver.page_source
    driver.quit()      
    return res

# Amazon の Seller ID を与えると， 販売者の情報を返す
def get_seller_detail( seller_id ):
    seller_url = 'https://www.amazon.co.jp/sp?_encoding=UTF8&seller=' + seller_id
    print( seller_url )
    bs = BeautifulSoup(get_page( seller_url ), features='lxml')
    data = bs.find_all('div',class_='a-row a-spacing-medium')[1]
    s = data.text
    ind1 = s.find("販売業者")
    ind2 = s.find("お問い合わせ先電話番号")
    ind3 = s.find("住所")
    ind4 = s.find("運営責任者名")
    ind5 = s.find("店舗名")
    #print( s[ind1+5:ind2] )
    #print( s[ind2+12:ind3] )
    #print( s[ind3+3:ind4] )
    #print( s[ind4+7:ind5] )
    #print( s[ind5+4:] )
    p1 = ""
    p2 = ""
    if len(  bs.select('.feedback-detail-description') ) > 0:
        point = bs.select('.feedback-detail-description')[0].text
        p_ind1 = point.find("過去12ヵ月間で")
        p_ind2 = point.find("評価：")
        p1 = point[ : p_ind2 -2 ]
        p2 = point[ p_ind2 +3: len(point)-1 ]
    return ( seller_id, p1, p2, s[ind1+5:ind2].replace('\n',''), s[ind2+12:ind3].replace('\n',''), s[ind3+3:ind4].replace('\n',''), s[ind4+7:ind5].replace('\n',''), s[ind5+4:].replace('\n',''), seller_url )
    #return ( seller_id, bs.select('#about-seller-text'),bs.select('.feedback-detail-description'), bs.select('#-component-heading') )

# Amazon の商品ページの URL を与えると，　その商品を販売している販売者の Seller ID の集合を返す
def get_seller_id_set( url ):
    seller_id_set = set([])
    bs = BeautifulSoup( get_page( url ), features='lxml' )
    links = bs.find_all('a') 
    for item in links:
        s = item.get('href')
        if ( type( s ) is str ) and ( 'seller=' in s ):
            index = s.find( 'seller' )
            if s[index+7] == 'A':
                seller_id_set.add( s[index+7:(index+21)] )
    return seller_id_set

# Amazon の商品検索ページの URL を与えると， 次のページの URL を返す ( いらない )
def get_next_page_url( url ):
    bs = BeautifulSoup( get_page( url ), features='lxml' )
    a_tag = bs.find('a', class_='s-pagination-item s-pagination-next s-pagination-button s-pagination-separator' )
    return "https://www.amazon.co.jp/" + a_tag.get('href')

# 商品検索ページ の URL を与えると， 商品ページの URL の集合を返す
def get_product_url_set( url ):
    bs = BeautifulSoup( get_page( url ), features='lxml' )
    a_tag_all = bs.find_all('a', class_='a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal' )
    product_url_set = set( [] )
    for item in a_tag_all:
        print( "-- 商品ページ URL 取得中 --" )
        product_url_set.add( "https://www.amazon.co.jp/" + item.get('href') )
    print( "商品ページ URL 取得終了" )
    return product_url_set

def write_csv( data ):
    with open('seller_id.txt', 'a') as f:
        print(data, file=f)
        
def write_csv_list( data ):
    with open('seller_info_ASAHIGOLF.txt', 'a') as f:
        print(','.join(data), file=f)
        
def main():
    count = 4
    print( "スタート" )
    url_manual = "https://www.amazon.co.jp/s?k=%E6%9C%9D%E6%97%A5%E3%82%B4%E3%83%AB%E3%83%95&__mk_ja_JP=%E3%82%AB%E3%82%BF%E3%82%AB%E3%83%8A&crid=1E9W1IYV23EJD&sprefix=%E6%9C%9D%E6%97%A5%E3%82%B4%E3%83%AB%E3%83%95%2Caps%2C300&ref=nb_sb_noss_"
    while True:
        if count > 5:
            break
        url_set = get_product_url_set( url_manual + str(count) )
        for item in url_set:
            sellers = get_seller_id_set( item )
            print( sellers )
            for item2 in sellers:
                write_csv( item2 )
                seller_data = get_seller_detail( item2 )
                write_csv_list( seller_data )
                print( seller_data  )
        count += 1

def main2():
    print( get_seller_detail( 'A29DF1EU4DWP7C' ) )

main()
```
